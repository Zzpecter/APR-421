# Calcular la Media

```python
median = 0.0
n = 0

n = len(data.steps)

ordered = data.steps.sort_values()


if n % 2 == 0:
    m1 = ordered.iloc[(n-2)//2]
    m2 = ordered.iloc[n//2]
    median = (m1 + m2) / 2
else:
    median = ordered.iloc[(n-2)//2]

median
```

# Calcular la Moda

```python
bucket_size = 1000
bucket_calculator = lambda x: int(x) / bucket_size * bucket_size

data['bucket'] = data.steps.apply(bucket_calculator)
bucket_min = data.bucket.min()
bucket_max = data.bucket.max()
bins = int((bucket_max-bucket_min)//bucket_size)
data.bucket.hist(bins=bins, figsize=(8,5), color='#00A99D', alpha=.5)

print(f"Mode: {data.groupby('bucket').steps.count().idxmax()}, with bucket size: {bucket_size}")
```
